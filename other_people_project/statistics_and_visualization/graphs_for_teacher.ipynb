# Graphs and Charts for Teacher

As an analyst I should create some graphs to help a high school math teacher. They want to display important metrics about their current class and the ones they have taught in previous years, in order to understand patterns and trends that may not be visible from just seeing the numbers in a spreadsheet.

In this project, I will crate some graphs using Matplotlib and a bit of NumPy.

## Visualization of Final Exam Averages via Bar Chart

On the graph are displayed average points for final exam by years with the margin of error. 

```python
from matplotlib import pyplot as plt

# List data for the graph
past_years_averages = [82, 84, 83, 86, 74, 84, 90]
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006]
error = [1.5, 2.1, 1.2, 3.2, 2.3, 1.7, 2.4]

# Draw barplot
plt.figure(figsize=(10,8))
plt.bar(range(len(past_years_averages)), past_years_averages, yerr = error, capsize = 5, color = 'Salmon')

# Zoom barplot
plt.axis([-0.5, 6.5, 70, 95])

# Add descriptions
ax = plt.subplot()
ax.set_xticks(range(len(years)))
ax.set_xticklabels(years)

plt.title('Final Exam Averages')
plt.xlabel('Year')
plt.ylabel('Test average')
plt.show()
```

## Average Scores via Side By Side Bars

This graph displays the differences in average test scores between students who went to two different middle schools before enrolling in the high school we are preparing graphs for.

```python
# List data for the graph
unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
middle_school_a = [80, 85, 84, 83, 86]
middle_school_b = [73, 78, 77, 82, 86]

# Calculate locations of bars on the x axis
def create_x(t, w, n, d):
    return [t*x + w*n for x in range(d)]

# Draw barplot
school_a_x = create_x(2, 0.8, 1, len(middle_school_a))
school_b_x = create_x(2, 0.8, 2, len(middle_school_b))
plt.figure(figsize = (10, 8))
plt.bar(school_a_x, middle_school_a, color = 'Violet', label = 'Middle School A')
plt.bar(school_b_x, middle_school_b, color = 'RebeccaPurple', label = 'Middle School B')

# Add descriptions
ax = plt.subplot()
middle_x = [a + b / 2 for a, b in zip(school_a_x, school_b_x)]
ax.set_xticks(middle_x)
ax.set_xticklabels(unit_topics)

plt.legend(loc=0)

plt.title('Test Averages on Different Units')
plt.xlabel('Unit')
plt.ylabel('Test Average')

plt.show()
```

## Break Down by Grades via Stacked Bars

The following graph displays the breakdown of students who got As, Bs, Cs, Ds, and Fs in each unit.


```python
import numpy as np

# List data for the graph
unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
As = [6, 3, 4, 3, 5]
Bs = [8, 12, 8, 9, 10]
Cs = [13, 12, 15, 13, 14]
Ds = [2, 3, 3, 2, 1]
Fs = [1, 0, 0, 3, 0]


# Calculate x axis and bottoms
x = range(len(unit_topics))
c_bottom = np.add(As, Bs)
d_bottom = np.add(c_bottom, Cs)
f_bottom = np.add(d_bottom, Ds)

# Draw stacked barplot
plt.figure(figsize=(10, 8))
plt.bar(x, As, label = 'A', color = 'Tomato')
plt.bar(x, Bs, bottom = As, label = 'B', color = 'Violet')
plt.bar(x, Cs, bottom = c_bottom, label = 'C', color = 'RebeccaPurple')
plt.bar(x, Ds, bottom = d_bottom, label = 'D', color = 'PaleVioletRed')
plt.bar(x, Fs, bottom = f_bottom, label = 'F', color = 'MediumSlateBlue')

# Add descriptions
ax = plt.subplot()
ax.set_xticks(x)
ax.set_xticklabels(unit_topics)

plt.title('Grade Distribution')
plt.xlabel('Unit')
plt.ylabel('Number of Students')

plt.legend(loc=0)
plt.show()
```

## Hardest topic

Now, we are going to look at the chart, it displays what students think the hardest topic covered throughout your math course is.

```python
# List data for the graph
unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
num_hardest_reported = [1, 3, 10, 15, 1]

# Draw piechart
plt.figure(figsize=(10, 8))
plt.pie(
    num_hardest_reported, 
    colors = ['Tomato', 'Violet', 'RebeccaPurple', 'PaleVioletRed', 'MediumSlateBlue'], 
    labels = unit_topics, 
    autopct = '%d%%', 
    textprops={'color':"w"},
)
plt.axis('equal')

# Add descriptions
plt.legend(unit_topics)

plt.title('Hardest Topics', color = 'White')

plt.show()
```

## Studying Time VS Exam Scores

The last graph displays the relationship between the time students say they studied and the scores they received on their final exams. We are going to use a lineplot with a shaded error.

```python
# List data for the graph
hours_reported =[3, 2.5, 2.75, 2.5, 2.75, 3.0, 3.5, 3.25, 3.25,  3.5, 3.5, 3.75, 3.75,4, 4.0, 3.75,  4.0, 4.25, 4.25, 4.5, 4.5, 5.0, 5.25, 5, 5.25, 5.5, 5.5, 5.75, 5.25, 4.75]
exam_scores = [52.53, 59.05, 61.15, 61.72, 62.58, 62.98, 64.99, 67.63, 68.52, 70.29, 71.33, 72.15, 72.67, 73.85, 74.44, 75.62, 76.81, 77.82, 78.16, 78.94, 79.08, 80.31, 80.77, 81.37, 85.13, 85.38, 89.34, 90.75, 97.24, 98.31]


# Get lower and upper bounds
hours_lower_bound = [0.8 * i for i in hours_reported]
hours_upper_bound = [1.2 * i for i in hours_reported]

# Draw lineplot
plt.figure(figsize=(10,8))
plt.plot(exam_scores, hours_reported, linewidth = 2, color = 'RebeccaPurple')
plt.fill_between(exam_scores, hours_lower_bound, hours_upper_bound, alpha = 0.2, color = 'RebeccaPurple')

# Add descriptions
plt.title('Time spent studying vs final exam scores')
plt.xlabel('Score')
plt.ylabel('Hours studying (self-reported)')

plt.show()
```
